```python
import sys
sys.path.append("../../../python")
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pilot4
import os
from utility import load_scene
```

```python
# Plotting parameters
sns.set_palette("tab10")
sns.set_style("whitegrid")
sns.set_context("talk")
sns.set(rc={'figure.figsize':(12,10)})
```

## Importing Data

```python
data = pd.read_csv("../data/data.csv")
```

```python
data = data.drop(columns=['success', 
                          'failed_images',
                          'failed_audio',
                          'failed_video',
                         'time_elapsed',
                         'scene_sp',
                         'internal_node_id',
                         'session_id',
                         'study_id',
                         'value',
                         'stimulus',
                         'timeout',
                         'trial_type'])
```

```python
# Convert all scene type entries to strings
data.scene_type = data.scene_type.astype(str)
# Retain only the scene type (remove "_goalpos_" substring)
data.scene_type = data.scene_type.apply(lambda x: "_".join(x.split('_')[:-2]))
```

```python
# Convert all scene entries to strings
data.scene = data.scene.astype(str)
# Retain only the scene name (remove ".mp4" substring)
data.scene = data.scene.apply(lambda x: x.split('.')[0])
```

```python
# List of subjects who failed comprehension check
failed_comprehension = []
# Iterate through all subjects
for subject in pd.unique(data.subject_id):
    # Find the number of attempts they made
    attempts = data.loc[(data.subject_id == subject) & (data.task == 'comprehension')].attempts.to_list()
    if max(attempts) > 3:
        failed_comprehension.append(subject)
print(f"{len(failed_comprehension)} of {len(pd.unique(data.subject_id))} ({15/42*100:.4}%) participants failed the comprehension check")
```

```python
# Remove people who failed comprehension check
data = data.loc[~(data.subject_id.isin(failed_comprehension))]
```

```python
# Check that none of the remaining subjects failed the comprehension check
l = pd.unique(data.subject_id)
all([x in failed_comprehension for x in l])
```

```python
# Copy dataframe
responses = data
```

```python
# Remove attempts column since we no longer are using it
responses = responses.drop(columns=['attempts'])
```

```python
# Remove all tasks that aren't coded as response
responses = responses.loc[(responses.task == 'response')]
```

```python
# Remove the task column since we no longer are using it
responses = responses.drop(columns=['task'])
```

```python
# Calculate z-scores
responses['zrt'] = (responses.rt - responses.rt.mean()) / responses.rt.std()
```

```python
# Remove outliers
responses = responses[abs(responses.zrt) <= 2]
```

```python
# Create dataframe of average rts per stimulus
stim_rt = pd.DataFrame(responses.groupby(['scene_type','scene','scene_index']).rt.mean())
stim_rt = stim_rt.reset_index()
```

```python
# Create dataframe of average rts per stimulus type
stim_type_rt = pd.DataFrame(stim_rt.groupby(['scene_type','scene_index']).rt.mean())
stim_type_rt = stim_type_rt.reset_index()
```

```python
g = sns.scatterplot(x='scene_index',y='rt',data=stim_type_rt,hue='scene_type')
g.set(xlabel='Goal Position\n(closer to ball -> farther from ball)', 
      ylabel='Mean RT',
     title="RT per goal indices across stimulus types")
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

```python
# Create plot
g = sns.catplot(x='scene_index',y='rt',
                data=responses, hue='scene_type',
                col='scene_col', kind="point",
                legend = False)
# Set labels
g.set(xlabel='Goal Index\n(closer -> farther)', 
      ylabel='Mean RT')
# Plot legend
plt.legend(title="Stimulus Type",
           bbox_to_anchor=(1.02, 1),
           loc='upper left', 
           borderaxespad=0)
# Plot title
g.fig.suptitle('RT according to distance from Ball')
plt.tight_layout()
```

```python
def save_plot(pure_df,abs_df,pure_df1,abs_df1,filename,save=False):
    scene_type = filename.split('/')[-1]
    # Set figure object
    fig, axs = plt.subplots(1,4,figsize=(24,6),gridspec_kw={'width_ratios': [1,1.5,1.5,1.5]})
    # Title subplots
    axs[0].title.set_text("Runtime per cumulative distance traveled from origin to Goal")
    axs[1].title.set_text("Predicted RT for pure simulation")
    axs[2].title.set_text("Predicted RT for abstraction")
    axs[3].title.set_text("Empirical RT per Goal Position")
    # Set subplots
    g=sns.scatterplot(x="x", 
                      y="y", 
                      hue="runtime",
                      s=200,
                      data=pure_df,
                      palette="coolwarm",
                      edgecolor="none",
                      ax=axs[0])
    g.set(ylim=(0, 1000),xlim=(0,800))
    g.invert_yaxis()
    
    g=sns.scatterplot(x="distance_cumulative",
                      y="runtime",
                      hue="runtime",
                      s=200,
                      data=pure_df1,
                      edgecolor="none",
                      palette="coolwarm",
                      ax=axs[1])
    g.set(ylim=(0, 400),xlim=(0,1200))
    
    g=sns.scatterplot(x="distance_cumulative",
                      y="runtime",
                      hue="runtime",
                      s=200,
                      data=abs_df1,
                      edgecolor="none",
                      palette="coolwarm",
                      ax=axs[2])
    g.set(ylim=(0, 400),xlim=(0,1200))
    
    g = sns.pointplot(x='scene_index',
                      y='rt',
                      data=responses.loc[(responses.scene_type == scene_type) & (responses.scene_col == "yes")],
                      hue='st',
                      s=200,
                      palette="coolwarm",
                      ax=axs[3])
    
    
    if save:
        plt.savefig(filename)
    else:
        plt.show()
```

```python
# Directory with relevant JSONs
loaddir = "/Users/lollipop/projects/physics_abstraction/data/json/pilot4/types/"
json_files = [pos_json for pos_json in os.listdir(loaddir) if pos_json.endswith('.json')]

for file in json_files:
    # Grab the full trace of ball's trajectory
    full_trace = pilot4.scene_trace(loaddir+file)
    
    # Grab RT prediction from abstraction model
    abs_full = pilot4.scene_rt(full_trace, loaddir+file, abstraction=True)
    # Grab RT prediction from simulation model
    sim_full = pilot4.scene_rt(full_trace, loaddir+file)
    
    # Grab the sparse trace of ball's trajectory
    sparse_trace = pilot4.get_points(full_trace)
    # Grab RT prediction from abstraction model
    abs_sparse = pilot4.scene_rt(sparse_trace, loaddir+file, abstraction=True)
    # Grab RT prediction from simulation model
    sim_sparse = pilot4.scene_rt(sparse_trace, loaddir+file)
    # Plot results
    save_plot(sim_full,abs_full,
              sim_sparse,abs_sparse,
              loaddir+file.split('.')[0])
```

```python
# Director with relevant JSONs
loaddir = "/Users/lollipop/projects/physics_abstraction/data/json/pilot4/trial/"
json_files = [pos_json for pos_json in os.listdir(loaddir) if pos_json.endswith('.json')]

simtime = {}

# Iterate through stimuli files
for file in json_files:
    # Scene name
    scene_name = file.split(".")[0]
    # Load the stimuli
    scene = load_scene(loaddir+file)
    # Instatiate the scene
    scene.instantiate_scene()
    # Run the scene
    scene.run_path(False,*(35,200,0.9))
    # Get simulation time
    st = scene.physics.tick
    simtime[scene_name] = st
```

```python
# Dataframe with simulation times
df = pd.DataFrame.from_dict(simtime, orient='index')
df = df.reset_index()
df = df.rename(columns = {'index':'scene',0:'st'})
```
