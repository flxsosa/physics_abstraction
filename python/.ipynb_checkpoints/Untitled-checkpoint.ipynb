```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mu = 0
sigma = 1
sigma2 = 0.5
sigma3 = 0.1
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma),color='b')
plt.plot(x, stats.norm.pdf(x, mu, sigma),color='b')
plt.xlim=[-5,5]
plt.ylim=[0,1]
plt.show()
```
