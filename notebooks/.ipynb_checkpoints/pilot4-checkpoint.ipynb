## Utilities

```python
def scene_trace(fname):
    # Get Ball position trace
    scene = load_scene(fname) # Load specific scene
    scene.instantiate_scene() # Instantiate objects in pymunk space
    scene.run(False) # Run simulation headless
    return scene.trace # Grab ball trace
```

```python
def scene_rt(trace,fname,abstraction=False,params=(35,200,0.9)):
    # Starting position
    origin = trace[0] 

    # Runtimes to reach point from origin
    runtimes = []

    # Euclidean distances from the origin per point in trace
    distance_from_origin = []

    # Cumulative distance traveled from origin under pure simulation
    distance_travelled = []
    dx = 0 # Change in distance
    last_point = origin
    
    # Iterate through points and compute runtimes and distances
    for idx,point in enumerate(trace):
        # Compute cumulative distance traveled
        curr_point = point
        dx += last_point.get_distance(curr_point)
        last_point = curr_point
        distance_travelled.append(dx)

        # Compute Euclidean distance from origin
        distance_from_origin.append(origin.get_distance(point))

        # Load scene
        scene = load_scene(fname)

        # Replace Goal at point
        for o in scene.objects:
            if o.name == "Goal":
                o.body.position = point

        # Simulate
        scene.instantiate_scene()
        if abstraction:
            scene.run_path(False,*params)
        else:
            scene.run(False)

        # Record runtime
        runtimes.append(scene.physics.tick)
        
    return pd.DataFrame({
            "x":[x[0] for x in trace],
            "y":[x[1] for x in trace],
            "runtime":runtimes,
            "distance":distance_from_origin,
            "distance_cumulative":distance_travelled})
```

```python
def get_points(trace,N=10):
    '''
    Returns a sublist of N evenly-spaced
    elements from the original list.
    
    :param trace: Original list to be indexed
    :param N: The distance between each index
    '''
    n = int(len(trace)/N)
    sublist = np.arange(0,len(trace),n) 
    if len(trace)-1 not in sublist:
        sublist = np.append(sublist,len(trace)-1)
    return [trace[i] for i in sublist]
```

```python
def save_plot(pure_df,abs_df,pure_df1,abs_df1,filename,save=True):
    # Set figure object
    fig, axs = plt.subplots(2,3,figsize=(30,20),gridspec_kw={'width_ratios': [1.75,1.75,1]})
    # Title subplots
    axs[0,0].title.set_text("Runtime per cumulative distance traveled from origin to Goal")
    axs[0,1].title.set_text("Runtime per (x,y) position of Goal")
    # Set subplots
    sns.scatterplot(x="distance_cumulative",y="runtime",hue="runtime",s=300,data=pure_df,edgecolor="none",palette="coolwarm",ax=axs[0,0])
    sns.scatterplot(x="distance_cumulative",y="runtime",hue="runtime",s=300,data=pure_df1,edgecolor="none",palette="coolwarm",ax=axs[0,1])
    g=sns.scatterplot(x="x", y="y", hue="runtime",s=400,data=pure_df,palette="coolwarm",edgecolor="none",ax=axs[0,2])
    g.set(ylim=(0, 1000),xlim=(0,800))
    g.invert_yaxis()

    # Title subplots
    axs[1,0].title.set_text("Runtime per cumulative distance traveled from origin to Goal")
    axs[1,1].title.set_text("Runtime per (x,y) position of Goal")
    # Set subplots
    sns.scatterplot(x="distance_cumulative",y="runtime",hue="runtime",s=300,data=abs_df,edgecolor="none",palette="coolwarm",ax=axs[1,0])
    sns.scatterplot(x="distance_cumulative",y="runtime",hue="runtime",s=300,data=abs_df1,edgecolor="none",palette="coolwarm",ax=axs[1,1])
    g=sns.scatterplot(x="x", y="y", hue="runtime",s=400,data=abs_df,palette="coolwarm",edgecolor="none",ax=axs[1,2])
    g.set(ylim=(0, 1000),xlim=(0,800))
    g.invert_yaxis()
    if save:
        plt.savefig(filename)
    else:
        plt.show()
```
